```python
import pandas as pd
```

```python
data = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
test_ids = test["PassengerId"]

data.head()
```

```python
# Clean our data set
def clean(data):
    data = data.drop(["PassengerId", "Name", "Ticket", "Cabin"], axis=1) # drop the unused cols
    # Fill NaN in cols
    cols = ["SibSp", "Age", "Parch", "Fare", "Pclass"]
    for col in cols:
        data[col].fillna(data[col].median(), inplace=True)
    # Deal with Nan in "Embarked" col
    data.Embarked.fillna("U", inplace=True)
    return data
# Run the function
data = clean(data)
test = clean(test)
```

```python
# Check or cleaned dataset
data.head()
```

```python
# Convert or string values to integers
from sklearn import preprocessing
# we will do a label encoder
le = preprocessing.LabelEncoder()
cols = [ "Sex", "Embarked"]
for col in cols:
    data[col] = le.fit_transform(data[col])
    test[col] = le.transform(test[col])
    print(le.classes_) 
data.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
y = data["Survived"]
x = data.drop("Survived", axis=1)
```

```python
model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(x, y)
```

```python
predictions = model.predict(test)
```

```python
df = pd.DataFrame({"PassengerId": test_ids.values,
                  "Survived": predictions,
                  })
```

```python
df.to_csv("Submission 2.csv", index=False)
```
